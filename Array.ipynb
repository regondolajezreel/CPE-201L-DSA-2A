```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("\nTraversing the array:\n")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")
def reverse(arr):
  print("\nReversing all elements:\n")
  reversed_arr = arr[::-1]
  print(reversed_arr)

#sort
def sort(arr):
  print("\nSorting all elements:\n")
  sorted_arr = sorted(arr)
  print(sorted_arr)

#insert
def insert(arr):
  print("\nInserting new element:\n")
  index, value = index, value = input("Enter index and value (use space): ").split()
  inserted_arr = arr
  inserted_arr.insert(int(index), int(value))
  print(inserted_arr)

#delete
def delete(arr):
  print("\nDeleting an element:\n")
  index = input("Enter an index to delete: ")
  deleted_arr = arr
  deleted_arr.pop(int(index))
  print(deleted_arr)

#modify
def modify(arr):
  print("\nModify an element:\n")
  index, value = index, value = input("Enter index and modified value (use space): ").split()
  modified_arr = arr
  modified_arr.pop(int(index))
  modified_arr.insert(int(index), int(value))
  print(modified_arr)




def main():
    # Initialize an array
    arr = [20, 10, 40, 50, 30]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse")
        print("2. Reverse")
        print("3. Sort")
        print("4. Insert")
        print("5. Delete")
        print("6. Modify")
        print("7. Exit")
        choice = input("Enter your choice (1-7): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reversed_arr = reverse(arr)
        elif choice == '3':
            sorted_arr = sort(arr)
        elif choice == '4':
            inserted_arr = insert(arr)
        elif choice == '5':
            deleted_arr = delete(arr)
        elif choice == '6':
            modified_arr = modify(arr)
        elif choice == '7':
            break;
        else:
          print("\nInvalid choice. Please enter a number between 1 to 7.")

if __name__ == "__main__":
    main()
```
